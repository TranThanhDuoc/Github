```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
```

```
# Initialize click-through rate and signup rate dictionaries
ct_rate = {'low':0.01, 'high':np.random.uniform(low=0.01, high=1.2*0.01)}
su_rate = {'low':0.2, 'high':np.random.uniform(low=0.2, high=1.2*0.2)}

def get_signups(cost, ct_rate, su_rate, sims):
    lam = np.random.normal(loc=100000, scale=2000, size=sims)
    # Simulate impressions(poisson), clicks(binomial) and signups(binomial)
    impressions = np.random.poisson(lam = lam)
    clicks = np.random.binomial(impressions, p=ct_rate[cost])
    signups = np.random.binomial(clicks, su_rate[cost])
    return signups, clicks

def get_revenue(signups):
    pur = []
    rev = []
    np.random.seed(123)
    for s in signups:
        # Model purchases as binomial, purchase_values as exponential
        purchases = np.random.binomial(s, p=0.1)
        pur.append(purchases)
        purchase_values = np.random.exponential(scale=10000, size = purchases)
        # Append to revenue the sum of all purchase values.
        rev.append(purchase_values.sum())
    return rev, pur
rd_rate=np.random.choice(['low','high'])
signup=get_signups(rd_rate, ct_rate, su_rate, 100)
revenue=get_revenue(get_signups(rd_rate, ct_rate, su_rate, 100)[0])
rangeplot=np.arange(len(signup[0]))
```

```
plt.figure(figsize=(10, 8))
sns.regplot(x=signup[0], y=revenue[0], line_kws={"color":"blueviolet","alpha":0.7,"lw":3},scatter_kws={"color":"red","alpha":0.3,"s":100}, )
sns.set(style="white", color_codes=False)
plt.title("Simulation: 100 samples", loc='center', fontsize=26, fontweight=0, color='m')
plt.xlabel("Number of sign-up in each sample", fontsize=18, fontweight=0, color='m')
plt.ylabel("Revenue in each sample", fontsize=18, fontweight=0, color='m')
```

```
plt.figure(figsize=(20, 6))
p1=plt.bar(rangeplot, signup[0], color='forestgreen', alpha = 0.9)
p2=plt.bar(rangeplot, revenue[1],color='r', alpha=0.9)
plt.legend(signup[0], 'Sign-up')
plt.title("Different between sign-ups and purchases", loc='center', fontsize=26, fontweight=0, color='m')
plt.xlabel("Number of sampling", fontsize=18, fontweight=0, color='m')
plt.ylabel("Number of turns", fontsize=18, fontweight=0, color='m')
plt.legend([p1, p2], ['signups','purchases'])
plt.show()
```
